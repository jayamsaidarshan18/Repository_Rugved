#Program 3

```python
num = input("Enter number: ")

increasing = True
decreasing = False

for i in range(1, len(num)):
    if increasing:
        if num[i] > num[i-1]:
            continue
        elif num[i] < num[i-1]:
            increasing = False
            decreasing = True
        else:
            print("Not a Hill Number")
            exit()
    else:
        if num[i] >= num[i-1]:
            print("Not a Hill Number")
            exit()

print("Hill Number")
```

#Program 4

```python
def selection_sort_string(s):
    arr = list(s)
    n = len(arr)

    for i in range(n):
        min_idx = i
        for j in range(i+1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]

    return "".join(arr)

print(selection_sort_string("python"))
```

#Program 5

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n-1)

print(factorial(5))  # 120
```

#Program 8

```python
s = input("Enter string: ")
n = int(input("Enter n: "))

if len(s) % n != 0:
    print("Division not possible")
else:
    parts = []

    for i in range(0, len(s), n):
        part = s[i:i+n]
        parts.append(part)

    same = True
    for part in parts:
        if part != parts[0]:
            same = False
            break

    if same:
        for part in parts:
            print(part)
    else:
        print("Sequence is not same")
```

#Program 9

```python
def caesar_cipher(text, shift):
    result = ""
    for ch in text:
        if ch.isalpha():
            base = ord('a') if ch.islower() else ord('A')
            result += chr((ord(ch) - base + shift) % 26 + base)
        else:
            result += ch
    return result

print(caesar_cipher("Hello", 3))  # Khoor
```

#18. Batting averages (Top 10)

```python
bat_avg = deliveries.groupby("batsman").agg(
    runs=("batsman_runs", "sum"),
    balls=("ball", "count")
)
bat_avg["average"] = bat_avg["runs"] / bat_avg["balls"]
bat_avg.sort_values("average", ascending=False).head(10)
```

#20. Total Matches vs Wins vs Win Rate

```python
total_matches = matches["team1"].value_counts() + matches["team2"].value_counts()
wins = matches["winner"].value_counts()

win_rate = (wins / total_matches) * 100
pd.DataFrame({
    "Matches": total_matches,
    "Wins": wins,
    "Win Rate": win_rate
}).fillna(0)
```
